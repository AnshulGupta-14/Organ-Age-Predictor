```python
import pandas as pd
import glob
from functools import reduce

files = glob.glob(r"C:\Users\anshu\Desktop\DataSets\Liver\*.tsv")

dfs = []

for f in files:
    df = pd.read_csv(f, sep="\t")

    mzml_col = [c for c in df.columns if c.endswith(".mzML")][0]
    
    sample = f.split("\\")[-1].replace(".tsv", "")

    tiny = df[["Protein.Group", mzml_col]].copy()
    
    dfs.append(tiny)

master = reduce(lambda left, right: pd.merge(left, right, on="Protein.Group", how="outer"), dfs)

# master.to_csv("MasterMatrix.tsv", sep="\t", index=False)

print(master.head())
print("Done! Master matrix created.")
```

```python
master.columns = [col.split('\\')[-1] for col in master.columns]
```
